## Compute the mean population density for each canton and rank the cantons from most dense to least dense.

```python
# We start by calculating the mean with groupby Canton (where param 'as_index=False' turns df2 into dataframe istead of Series) and sort the data in descending order

df2 = (
    df.groupby(["Canton"], as_index=False)["Population density per km²"]
    .mean()
    .sort_values(["Population density per km²"], ascending=False)
)
df2["Rank"] = (
    df2["Population density per km²"].rank(ascending=False).astype("int")
)  # inserting Rank column with type 'int' for better presentation
df2 = df2.set_index(
    "Rank"
)  # another cosmetic change for better presentation by passing Rank values to index
df2
```

```python
# We could also use a shorter code to show a series of Cantons by rank only

df.groupby(["Canton"])["Population density per km²"].mean().rank(
    ascending=False
).sort_values()
```

## Compute the number of communes in each canton where more than 50 percent of their populations are aged between 20 and 64 years old.

```python
# here the approach is the same: we extract the necessary data from our initial df with the help of groupby, but this time by apply method with lambda func
# which will calculate the desired percentage. We use 'reset_index' at the end instead of groupby param 'as_index=False' for Series conversion into df.

df3 = (
    df.groupby(["Canton", "Commune"])
    .apply(lambda x: (x["20-64 years"] / x["Residents"]).sum())
    .reset_index(name="Communes >50%")
)
df3.head(10)
```

```python
# now we need to count the number of Commune lines >50% per Canton; again, apply with lambda comes in handy

df4 = (
    df3.groupby(["Canton"])["Communes >50%"]
    .apply(lambda x: (x > 0.5).sum())
    .reset_index()
)
df4[df4["Communes >50%"] > 0].sort_values(
    ["Communes >50%"], ascending=False
).reset_index(
    drop=True
)  # finally, we filter and sort the output with drop=True to replace index
```

# Probabilities

## Compute the probability that a randomly selected commune with elevation over 2000 is from the canton of Valais.

```python
# let's call the event (>2k) as A and the event (=VS) as B. In this case, we can conclude that the probability of event B is conditioned on event A or P(B| A).
# From the course we learned that P(B| A) = P(B). Thus, we can describe P(B) as follows:

len(df[(df.Canton == "VS") & (df.Elevation > 2000)]) / len(
    df[df.Elevation > 2000]
)  # 49/96
```

```python
# let's check this with the randomizer
import random

all_B = list(df.Canton[(df.Elevation > 2000)])  # list of all elevations >2k (96)
experiments = 10000
counter = 0  # counter of successfull B
for i in range(experiments):
    canton = all_B[
        random.randrange(len(all_B))
    ]  # we pass a random canton value from the list to the variable
    if canton == "VS":  # and check if it correspons to the condition
        counter += 1  # if it does, we count it as successfull

counter / experiments  # P(B)
```

## Define a data frame matrix whose rows correspond to communes and the columns to the cantons. Fill in the matrix with 0/1 values where entry (i,j) is a 1 if the commune in row i is in the canton in column j and a 0 otherwise.

```python
# my first trial on this exercies started with quite simple approach:
# create a copy and fill (i:j) with a nested 'for loop'

df7 = pd.DataFrame(
    df, df["Canton"], df.Canton.unique()
)  # using unique() to transpose unique cantons to columns
```

```python
for j in range(len(df7.columns)):
    for i in range(len(df7)):
        if df7.index[i] == df7.columns[j]:
            df7.iloc[[i], [j]] = 1
        else:
            df7.iloc[[i], [j]] = 0

df7 = df7.astype(int).copy()  # adding int and vaidate changes to df with copy()
df7.head(10)
# nothing special apart of noticeable latency of 'for loop' which will be even worse with bigger frames...
```
